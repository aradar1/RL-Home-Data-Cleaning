```python
import pandas as pd
from pathlib import Path
```

```python
df_reference_data = pd.read_csv('888 LETSFLO ASSETS.csv')
```

```python
df_888_data = pd.read_excel('888 Data.xlsx')
```

```python
#Clean Barcode for merge preparation, remove ="
#df['column_name'] = df['column_name'].str.replace(r'[abc]', '', regex=True)

df_reference_data['BARCODE'] = df_reference_data['BARCODE'].str.replace(r'="','',regex=True)
```

```python
#Remove ending in "
df_reference_data['BARCODE'] = df_reference_data['BARCODE'].str.replace('"','',regex=True)
```

```python
#Remove beginning with 00
df_reference_data['BARCODE'] = df_reference_data['BARCODE'].str.replace('^00','',regex=True)
```

```python
#Remove 'Home-' from Lighting Type
df_reference_data['TYPE'] = df_reference_data['TYPE'].str.replace('Home-','',regex=True)
df_reference_data.head()
```

```python
#remove null column
df_888_data.drop(df_888_data.columns[df_888_data.columns.str.contains('Unnamed', case=False)], axis=1, inplace=True)
```

```python
#df['column_name'] = df['column_name'].astype(str)
df_888_data['Receiving Report #'] = df_888_data['Receiving Report #'].astype(str)
```

```python
#Remove .0 on values in Barcode
df_888_data['Receiving Report #'] = df_888_data['Receiving Report #'].str.replace('.0','',regex=False)
```

```python
#df = df.rename(columns={'col1': 'new_col1', 'col2': 'new_col2'})
df_888_data = df_888_data.rename(columns={'Receiving Report #': 'BARCODE'})
```

```python
df_888_data = df_888_data[df_888_data['BARCODE'] != 'nan']
```

```python
df_888_data
```

```python
## Merge with an indicator column
#diff = pd.merge(df1, df2, on=['id', 'value'], how='outer', indicator=True)
# Filter rows where the indicator is not 'both'
#differences = diff[diff['_merge'] != 'both']
#print(differences)
diff_df = pd.merge(df_reference_data, df_888_data, on=['BARCODE'], how='outer', indicator=True)
dif = diff_df[diff_df['_merge'] !='both']
print(dif)
```
