```python
import pandas as pd
from pathlib import Path
```

```python
df_reference_data = pd.read_csv('888 LETSFLO ASSETS.csv')
```

```python
#Clean Barcode for merge preparation, remove ="
#df['column_name'] = df['column_name'].str.replace(r'[abc]', '', regex=True)

df_reference_data['BARCODE'] = df_reference_data['BARCODE'].str.replace(r'="','',regex=True)
```

```python
#Remove ending in "
df_reference_data['BARCODE'] = df_reference_data['BARCODE'].str.replace('"','',regex=True)
```

```python
#Remove beginning with 00
df_reference_data['BARCODE'] = df_reference_data['BARCODE'].str.replace('^00','',regex=True)
```

```python
#Remove 'Home-' from Lighting Type
df_reference_data['TYPE'] = df_reference_data['TYPE'].str.replace('Home-','',regex=True)
df_reference_data.head()
```

```python
df_reference_data
```

```python
df_888_data = pd.read_csv('888 Data.csv')
```

```python
#remove null column
df_888_data.drop(df_888_data.columns[df_888_data.columns.str.contains('Unnamed', case=False)], axis=1, inplace=True)
```

```python
#df['column_name'] = df['column_name'].astype(str)
df_888_data['Receiving Report #'] = df_888_data['Receiving Report #'].astype(str)
```

```python
#Remove .0 on values in Barcode
df_888_data['Receiving Report #'] = df_888_data['Receiving Report #'].str.replace('.0','',regex=False)
```

```python
#df = df.rename(columns={'col1': 'new_col1', 'col2': 'new_col2'})
df_888_data = df_888_data.rename(columns={'Receiving Report #': 'BARCODE'})
```

```python
df_888_data
```

```python
#Merged dataframes for matching data
matching_df = pd.merge(df_reference_data,df_888_data, on='BARCODE',how='inner')
```

```python
matching_df.head()
```

```python
matching_df
```

```python
#Merge dataframes for unmatching data
different_df = pd.merge(df_reference_data,df_888_data, on='BARCODE',how='left')
```

```python
different_df
```

```python
different_df_cleaned = different_df.dropna(subset=['DESCRIPTION'])
different_df_cleaned
```

```python
print(different_df_cleaned['BARCODE'].nunique())
```

```python
# Filter duplicate values from 'column_name'
#duplicates = df[df['column_name'].duplicated(keep=False)]
duplicates = different_df_cleaned[different_df_cleaned['BARCODE'].duplicated(keep=False)]
print(duplicates)
```

```python
duplicates
#Noted that same barcode but different SKUs, Different Retail/Cost
```

```python
#Make 888 Data as reference
home_df = pd.merge(df_reference_data,df_888_data, on='BARCODE',how='right')
home_df
```

```python
home_df = home_df.dropna(subset=['DESCRIPTION'])
home_df
```

```python
print(home_df['BARCODE'].nunique())
```

```python
#Print dupes
dupes = home_df[home_df['BARCODE'].duplicated(keep=False)]
dupes
```

```python
#Make Excel Spreadsheet using from LetsFlo data as reference merge
different_df_cleaned.to_csv('Output/output.csv',index=False)
```

```python
#Display differences in both 
concat_df = pd.concat([different_df_cleaned,home_df]).drop_duplicates(keep=False)
```
